```python
import os

import numpy as np
import pandas as pd
import geopandas as gpd

import rasterio
from rasterio.plot import show
from rasterstats import zonal_stats

import matplotlib.pyplot as plt
```

```python
gdf = gpd.read_file("gs://etrace-data/data/raw_data/NUTS_RG_01M_2024_3035_LEVL_2.geojson")
```

```python
# Calculate area in square meters (requires projected CRS)
gdf['area_m2'] = gdf.geometry.area

# Convert to square kilometers
gdf['area_km2'] = gdf['area_m2'] / 1_000_000

# Display a sample
print(gdf[['NUTS_ID', 'NUTS_NAME', 'area_km2']].head())
```

```python
# Export specific columns to CSV, excluding the index
gdf.to_csv('gs://etrace-data/data/raw_data/nuts2_areas.csv', columns=['NUTS_ID', 'NUTS_NAME', 'area_km2'], index=False)
```
